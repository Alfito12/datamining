## **Metode Decision Tree**

````python
import pandas as pd
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
````

````python
url = 'https://raw.githubusercontent.com/Alfito12/dataset/main/dataR2.csv'
data = pd.read_csv(url)
````

````python
data.head(-1)
````

````python
# x sebagai variable independent
x = data.drop(['Classification'], axis = 1)
x.head
````

````python
# y sebagai variable dependent
y = data['Classification']
y.head
````

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, shuffle=True, stratify=y, random_state=0)
````

````python
print(x_train)
````

````python
# membangun decision tree menggunakan gini
clf = DecisionTreeClassifier(criterion="gini")
clf = clf.fit(x_train,y_train)

#Prediksi data test
y_pred = clf.predict(x_test)
````

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
````

````python
plt.figure(figsize=(15,15))
#create the tree plot
a = tree.plot_tree(clf,
                   rounded = True,
                   filled = True,
                   fontsize=8)
#show the plot
plt.show()
````
